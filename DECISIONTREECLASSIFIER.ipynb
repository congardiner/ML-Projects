```
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import numpy as np

# Label Encoder is a sklearn function that can essentially map or assign values that are categorical like month or days, and assign them an int value
# This is just something cool that I learned that I wanted to test and further theorize, however, it might be better to just assign key value pairs.

# I had started with the decisiontreeregressor, however, after deeper research and findings, I determined that a decisiontreeclassifier might actually prove to be a better model as it would assess categorical data more efficiently rather than numerical
# Numerical data would simply require more legwork to get up and running for what we are working with instance wise.

## NEED to evaluate unbalanced and overfitting for the data exploration phase
# In addition to this, the data needs to be one-hot-encoded for 0,1 (ie, or another format so that the data itself isn't biased)
# Oversampling versus sampling; the dataset appears to be completely unbalanced, what would be the next focus? (Internal Question brought-up in our discussion)
# NEED to use a balanced weight (ie, what does this do?) ~ prevents one class or feature overdominating the model and providing inaccurate emphasis on one set of trained data.
# NEED to explore the data evenmore so that we can assemble what we need to... executive summary wise

bank_data = holdout_data.dropna().copy()

housing_mapping = {
    'yes': 1,
    'no': 0,
    'unknown': 2
}

bank_data['housing'] = bank_data['housing'].map(housing_mapping)

month_mapping = {
    'jan': 0,
    'feb': 1,
    'mar': 2,
    'apr': 3,
    'may': 4,
    'jun': 5,
    'jul': 6,
    'aug': 7,
    'sep': 8,
    'oct': 9,
    'nov': 10,
    'dec': 11
}

bank_data['month'] = bank_data['month'].map(month_mapping)


bank_data = bank_data[bank_data['poutcome'] != 'nonexistent']


# features = ['age', 'default', 'education', 'pdays', 'poutcome', 'cons.conf.idx']
# X = pd.get_dummies(bank_data[features], drop_first=True)
# y = bank_data['contact']

# features = ['age', 'default', 'education']
# X = pd.get_dummies(bank_data[features], drop_first=True)
# y = bank_data['dayofweek']

features = ['age', 'education', 'pdays', 'cons.conf.idx', 'nr.employed']
X = pd.get_dummies(bank_data[features], drop_first=True)
y = bank_data['poutcome'].map({'success': 1, 'failure': 0})

# label_encoder = LabelEncoder()
# y = label_encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)

clf = DecisionTreeClassifier(max_depth=3)

clf.fit(X_train, y_train)

clf.score(X_test, y_test)




```

```
y_pred = clf.predict(X_test)

conf_max = confusion_matrix(y_test, y_pred)

conf_max
```

```
X_encode = pd.get_dummies(X, drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(X_encode, y, test_size=0.2, random_state=42)

model = DecisionTreeClassifier(max_depth=3)
model.fit(X_train, y_train)


feature_importances = pd.DataFrame({
    'feature': X_encode.columns,
    'importance': model.feature_importances_
}).sort_values('importance', ascending=False)

print(feature_importances.head(20))
```
